# **Import Library**

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

import seaborn as sns
```

# **Upload the Data**

```python
df =pd.read_csv('/content/titanic_dataset.csv')[['Age', 'Pclass', 'SibSp','Parch', 'Survived']]
```

```python
df.head()
```

# **Check mising value**

```python
df.isnull().sum()
```

# **Drop Mssing value**

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
X = df.iloc[:, 0:4]
y =df.iloc[:,-1]
```

```python
X.head()
```

```python
y.head()
```

# **Cross_value_Score**

```python
np.mean(cross_val_score(LogisticRegression(),X,y,scoring='accuracy', cv=20))
```

# **Apply Feature Construction**

```python
X['family_size']= X['SibSp'] + X['Parch'] +1
```

```python
X.head()
```

```python
def myfunc(num):
  if num ==1:
    #alone
    return 0
  elif num>1 and num <=4:
    #small  family
    return 1
  else:
    #large family
    return 2
```

```python
myfunc(2)
```

```python
X['Family_type'] = X['family_size'].apply(myfunc)
```

```python
X.head()
```

# **Drop Columns**

```python
X.drop(columns=['SibSp', 'Parch', 'family_size'], inplace=True)
```

```python
X.head()
```

```python
np.mean(cross_val_score(LogisticRegression(), X,y ,scoring='accuracy', cv=20))
```
